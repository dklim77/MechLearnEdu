```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
data = pd.read_excel("D:\MAGVI_500RT_RawData.xlsx")
data.info()
```

```python
print(data.shape)
data[:5]
```

```python
# 사본 사용 (원본 백업)
df = data.copy()
df[:3]
```

```python
X = df[["EvaporatingTemp", "CondensingTemp", "IGVOpening", "RPM"]]
y = df[["Capacity", "InputPower", "FirstTopHead", "SecondTopHead", "1stCompHead", "2ndCompHead"]]
X[:5], y[:5]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=9)
model = RandomForestRegressor(n_estimators=20)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
X_test.shape, y_pred.shape
```

```python
model.fit(X_train, y_train["Capacity"])
R_Square_Capacity = model.score(X_test, y_test["Capacity"])

model.fit(X_train, y_train["InputPower"])
R_Square_InputPower = model.score(X_test, y_test["InputPower"])

model.fit(X_train, y_train["FirstTopHead"])
R_Square_FirstTopHead = model.score(X_test, y_test["FirstTopHead"])

model.fit(X_train, y_train["SecondTopHead"])
R_Square_SecondTopHead = model.score(X_test, y_test["SecondTopHead"])

model.fit(X_train, y_train["1stCompHead"])
R_Square_1stCompHead = model.score(X_test, y_test["1stCompHead"])

model.fit(X_train, y_train["2ndCompHead"])
R_Square_2ndCompHead = model.score(X_test, y_test["2ndCompHead"])

print("R_Square_Capacity : ", R_Square_Capacity)
print("R_Square_InputPower : ", R_Square_InputPower)
print("R_Square_FirstTopHead : ", R_Square_FirstTopHead)
print("R_Square_SecondTopHead : ", R_Square_SecondTopHead)
print("R_Square_1stCompHead : ", R_Square_1stCompHead)
print("R_Square_2ndCompHead : ", R_Square_2ndCompHead)
```
